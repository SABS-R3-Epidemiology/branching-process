```python
# Import libraries
import branchpro
import nbformat
import numpy as np
import plotly.graph_objects as go

print(nbformat.__version__)
```

```python
# Read french-flu data
data = branchpro.DatasetLibraryAPI().french_flu()

french_flu_plot = branchpro.IncidenceNumberPlot()
french_flu_plot.add_simulation(data, time_key='time_index', inc_key='inc')
french_flu_plot.show_figure()
```
